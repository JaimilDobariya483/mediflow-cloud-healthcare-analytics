### Dataset 

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 10000

data = {
    "patient_id": range(1, n+1),
    "age": np.random.randint(18, 90, n),
    "gender": np.random.choice(["Male", "Female"], n),
    "department": np.random.choice(
        ["Cardiology", "Emergency", "Neurology", "Orthopedics", "ICU"], n),
    "disease": np.random.choice(
        ["Heart Disease", "Stroke", "Fracture", "Infection", "Diabetes"], n),
    "length_of_stay": np.random.randint(1, 15, n),
    "readmitted": np.random.choice([0,1], n, p=[0.8,0.2]),
    "hospital_city": np.random.choice(
        ["New York","Chicago","Dallas","Los Angeles","Houston"], n)
}

df = pd.DataFrame(data)

df.to_csv("../data/raw/hospital_data.csv", index=False)

print("Dataset Created Successfully")
```
